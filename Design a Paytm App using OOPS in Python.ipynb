```python
"""

Create a Paytm App by using Core Python (oops) it include- 
app name, 
wallet amount by default is zero, 
user and admin registration and updation- raise exception if anyone try to delete admin credentials ,  
user-login, signup, top-up(adding money to wallet), recharge  mobile

"""



"""  Only For Single User Approach  """


class PaymentApp():
    
    def __init__(self,name,adminid,passwd):
        self._appName = name
        self._adminID=adminid
        self._adminPasswd=passwd
        self.user={}
        self.wallet=0

           
    def user_reg(self,email, passwd):
        self._userEmail = email
        self._userPasswd = passwd
        self.user={"ID":self._userEmail,"PAS":self._userPasswd}
        return "\n!! Sign-Up Successfully !!\n"
     
    def user_login(self, email, passwd):
        if len(self.user)!=0:             #or len(self.user.keys())
            if email==self.user["ID"] and passwd==self.user["PAS"]:
                return "!! Login Successfully !!"
            else:
                return "!! Incorrect Details !!"
        else:
            return "!! Incorrect Details !!"    
        
    def adminlogin(self, userid, passwd):
        if userid == self._adminID and passwd ==self._adminPasswd :
            return "!! Login Successfully !!"
        
        else:
            return "!! Incorrect Details !!"
        
        # getter method for user email    
    @property
    def userEmail(self):
        return self._userEmail
     
    # getter method for user pass
    @property
    def userPasswd(self):
        return self._userPasswd
    
    # setter method for user email
    @userEmail.setter
    def userEmail(self, new_email):
        if new_email!=self.user["ID"]:
            self.user["ID"]=new_email
            print("\n!! Username Updated Successfully !!\n")
        else:
            print("\n!! New Username can't be same as old username !!\n")     
    
    # setter method for user pass
    @userPasswd.setter
    def userPass(self, new_pass):
        if new_pass!=self.user["PAS"]:
            self.user["PAS"]=new_pass
            print("\n!! Password Updated Successfully !!\n")
        else:
            print("\n!! New Password can't be same as old password !!\n")   
    
    # getter method for admin id
    @property
    def adminID(self):
        return self._adminID  
    
    # getter method for admin password
    @property
    def adminPasswd(self):
        return self._adminPasswd
      
    # getter method for user email
    @adminID.setter
    def adminID(self, new_adminid):
        if new_adminid!=self._adminID:
            self._adminID=new_adminid
            print("\n!! Admin ID Updated Successfully !!\n")
        else:
            print("\n!! New ID can't be same as old ID !!\n")
        
    # getter method for user email
    @adminPasswd.setter
    def adminPasswd(self, new_adminpasswd):
        if new_adminpasswd!=self._adminPasswd:
            self._adminPasswd=new_adminpasswd
            print("\n!! Password Updated Successfully !!\n")
        else:
            print("\n!! New Password can't be same as old password !!\n")
     
    # deleter method for admin id
    @adminID.deleter
    def adminID(self):
        raise Exception ("Admin Details can't be delete")
    
    # getter method for app name
    @property
    def appName(self):
        return self._appName
    
    # setter method for app name
    @appName.setter
    def appName(self, new_name):
        if new_name!=self._appName:
            self._appName=new_name
            print("\n!! App Name Updated Successfully !!\n")
        else:
            print("\n!! New Name can't be same as old name !!\n")
    
    
    def addmoney(self, amount):
        if amount>0:
            self.wallet+=amount
            print("\n!! Amount added Successfully !!\n")
        else:
            print("\n!! Amount 0.0 Can't be added\n")
            
    def recharge(self, amount):
        if amount>0 and amount<=self.wallet:
            self.wallet-=amount
            print("\n!! Recharge Done !!\n")
        else:
            print("\n!! Insufficient Wallet Balance Please Add Money First !!\n")
      
    
if __name__=='__main__':
    try:
        obj=PaymentApp("PAYTM","123","123")
        print(f"\nWELCOME TO {obj.appName} PAYMENT BANK OF INDIA\n\n")
        print("USE THIS FOR RUNNING MY CODE : (Admin ID- 123, Admin PASSWORD- 123)" )
        while True:
            print("\n1. ADMIN LOGIN")
            print("2. USER")
            print("3. EXIT\n")
            x=input()
            if x=="1":
                userid=input('Enter Admin ID : ')
                passwd=input('Enter Password : ')
                res=obj.adminlogin(userid, passwd)
                if res=="!! Login Successfully !!":
                    print("\n!! Login Successfully !!")
                    while True:
                        print("\nADMIN DETAILS : \n\nAdmin ID :", obj.adminID)
                        print("Admin Password :", obj.adminPasswd)
                        print('\n1. Update Admin ID')
                        print('2. Update Admin Password')
                        print('3. Delete Admin Account')
                        print("4. Update App Name")
                        print("5. Exit\n")
                        x=input()
                        if x=="1":
                            new_adminid=input("Enter New Admin ID : ")
                            obj.adminID=new_adminid
                        elif x=="2":
                            new_adminpasswd=input("Emter New Admin Password : ")
                            obj.adminPasswd=new_adminpasswd
                        elif x=="3":
                            del obj.adminID
                        elif x=="4":
                            print("App Name : \n",obj.appName,"\n")
                            new_name=input("Enter New App Name :")
                            obj.appName=new_name
                        elif x=="5":
                            break
                        else:
                            print("\n!! Please Enter the valid input !!\n")
                else:
                    print("\n!! Please Enter the Valid Credentials !!\n")
                    
            elif x=="2":
                while True:
                    print("\n1. User Registration")
                    print("2. User Login")
                    print("3. Exit\n")
                    y=input()
                    if y=="1":
                        email=input("Enter your email ID : ")
                        passwd=input('Enter your password : ')
                        res=obj.user_reg(email, passwd)
                        print(res)
                        
                    elif y=="2":
                        email=input("Enter your email ID : ")
                        passwd=input('Enter your password : ')
                        res=obj.user_login(email, passwd)
                        if res=="!! Login Successfully !!":
                            print("\n!! Login Successfully !!")
                            while True:
                                print("\n1. Add Money\n2. Recharge\n3. Wallet Balance\n4. Profile Details\n5. Exit")
                                x=input()
                                if x=="1":
                                    amount=int(input("Enter the Amount : "))
                                    obj.addmoney(amount)
                                    print("UPDATED WALLET : ",obj.wallet)
                                    
                                elif x=='2':
                                    amount=int(input("Enter the Amount : "))
                                    obj.recharge(amount)
                                    print("UPDATED WALLET : ",obj.wallet)
                                
                                elif x=="3":
                                    print("\nWALLET BALANCE : ",obj.wallet)
                                    
                                elif x=="4":
                                    while True:
                                        print("\n1. Update User ID\n2. Update Password\n3. Exit")
                                        x=input()
                                        if x=="1":
                                            new_email=input("Enter new Username : ")
                                            obj.userEmail=new_email
                                            print("\nUPDATED USER NAME :", obj.userEmail)
                                            
                                        elif x=="2":
                                            new_pass=input("Enter new password : ")
                                            obj.userPass=new_pass
                                            print("UPDATED USER PASSWORD :", obj.userPasswd)
                                        elif x=="3":
                                            break
                                        else:
                                            print("\n!! Please enter the correct input !!\n")
                                        
                                elif x=="5":
                                    break
                                else:
                                    print("\n!! Please enter valid input !!\n")         
                        else:
                            print("\n!! Please Enter the Valid Credentials !!\n")
                            
                    elif y=="3":
                        break
                    else:
                        print("!! Please Enter the valid input !!\n")
            
            elif x=="3":
                break
            else:
                print("\nInvalid Input\n")
    except Exception as e:
        print(e)
        print('\nPLEASE PROVIDE VALID INPUT\n')
print("!! \nTHANK YOU !!")                                                  
```
